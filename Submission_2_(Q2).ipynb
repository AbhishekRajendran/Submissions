Insights from Data for 'Sikka' and 'Sikka Pro' Apps:

To compare the performance of the two apps, we can look at the following data points:

Offer Initiation by Users:
We can analyze the number of offers initiated by users on each app. This will give us an idea of which app is more popular and engaging for users.

Offer Completion by Users:
We can look at the number of offers completed by users on each app. This will give us insights into the user engagement and experience on each app.

Rewards Earned by Users:
We can analyze the total rewards earned by users on each app. This will give us 
 idea of which app is more rewarding for users and therefore, more likely to retain them.

Revenue Generated:
We can analyze the total revenue generated by each app. This will give us insights into the profitability and success of each app.

To perform this analysis, we will use the following tables:

User Signup Data: This table will provide us with the details of every user coming through different marketing channels on ‘Sikka’ or ‘Sikka Pro’ app.

User Offer Data: This table will provide us with the details of every offer initiated on ‘Sikka’ or ‘Sikka Pro’ app.

User Offer Completion Data: This table will provide us with the data of the offers completed by any ‘Sikka’ or ‘Sikka Pro’ user.

Rewards Details: This table will provide us with the details of every offer-reward that is in the ‘Sikka’ or ‘Sikka Pro’ App.

Using the above data points and tables, we can perform the following analysis to compare the performance of the two apps:

```python
import pandas as pd
```

```python
import pandas as pd
from datetime import datetime

# Load the user signup data
user_signup_data = pd.read_csv('/content/Q2_users signup.csv')

# Load the user offer completion data
user_offer_completion_data = pd.read_csv('/content/Q2_User offer completion data (1).csv')

# Load the rewards details data
rewards_details_data = pd.read_csv('/content/Q2_rewards details.csv')

```

```python
# Load the User offer data 
user_offer_data = pd.read_csv('/content/Q2_User offer data.csv')
```

```python
# Merge user signup data with user offer data
user_data = pd.merge(user_signup_data, user_offer_data, on='user_id')
```

```python
user_data['status'].unique()
```

```python
# Find offer initiation by users for each app
```

```python
offer_initiation = user_data.groupby('app_id')['offer_id'].count()
print("Offer Initiation by Users:")
print(offer_initiation)
```

```python
import matplotlib.pyplot as plt

# plot the bar chart
fig, ax = plt.subplots(figsize=(10, 8))
offer_initiation.plot(kind='bar', color=['orange', 'brown'], ax=ax)

# set the title and labels
ax.set_title('Offer Initiation by Users', fontsize=16)
ax.set_xlabel('App ID', fontsize=12)
ax.set_ylabel('Number of Offers Initiated', fontsize=12)

# set the color and background
ax.set_facecolor('beige')
ax.spines['bottom'].set_color('gray')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# add count labels on top of each bar
for i in ax.containers:
    ax.bar_label(i, label_type='edge')
    
# show the plot
plt.show()
```

```python
# Find offer completion by users for each app
```

```python
Offer_completion = user_offer_completion_data.groupby('app_id')['user_id'].count()
print("Offer_completion by Users:")
print(Offer_completion)
```

```python
import matplotlib.pyplot as plt

# plot the bar chart
fig, ax = plt.subplots(figsize=(10, 8))
Offer_completion.plot(kind='bar', color=['orange', 'brown'], ax=ax)

# set the title and labels
ax.set_title('Offer_completion by Users', fontsize=16)
ax.set_xlabel('App ID', fontsize=12)
ax.set_ylabel('Number of Offers Initiated', fontsize=12)

# set the color and background
ax.set_facecolor('beige')
ax.spines['bottom'].set_color('gray')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# add count labels on top of each bar
for i in ax.containers:
    ax.bar_label(i, label_type='edge')
    
# show the plot
plt.show()
```

**Rewards Earned By Users**

```python
# Join the user offer completion data with the rewards details data
user_rewards_data = pd.merge(user_offer_completion_data, rewards_details_data, on='reward_id')
```

```python
# Convert total_payout_in_paise to rupees
user_rewards_data['total_payout_in_rupees'] = user_rewards_data['total_payout_in_paise'] / 100
```

```python
# Group the data by app_id and calculate the sum of total_payout_in_paise
total_rewards_by_app = user_rewards_data.groupby('app_id')['total_payout_in_rupees'].sum()

print(total_rewards_by_app)
```

```python
import matplotlib.pyplot as plt


# Plot a pie chart for the total rewards earned by each app
plt.pie(total_rewards_by_app, labels=total_rewards_by_app.index, autopct='%1.1f%%', colors=['#7fc97f','#beaed4'])

# Set the title of the pie chart
plt.title('Total Rewards Earned by App')

# Show the pie chart
plt.show()

```

**Revenue Generated**

```python
# Join the user offer completion data with the rewards details data
user_rewards_data = pd.merge(user_offer_completion_data, rewards_details_data, on=['reward_id'])
```

```python
# Convert total_revenue to rupees
user_rewards_data['total_revenue_in_rupees'] = user_rewards_data['total_revenue_in_paise'] / 100
```

```python
# Group the data by app_id and sum the total revenue
revenue_data = user_rewards_data.groupby('app_id')['total_revenue_in_rupees'].sum()

# Print the revenue data
print(revenue_data)
```

```python
import matplotlib.pyplot as plt


# Plot a pie chart for the total revenue earned by each app
plt.pie(revenue_data, labels=revenue_data.index, autopct='%1.1f%%', colors=['#7fc97f','#beaed4'])

# Set the title of the pie chart
plt.title('Total Revenue Earned by App')

# Show the pie chart
plt.show()

```

Based on the above calculations, the following insights can be drawn for the 'Sikka' and 'Sikka Pro' apps:

Offer initiation by users is slightly higher on 'Sikka' app as compared to 'Sikka Pro' app. However, the difference is not significant.

The number of offers completed by users is higher on 'Sikka Pro' app as compared to 'Sikka' app, which indicates better user engagement and experience on 'Sikka Pro' app.

Although the difference is not significant, the total rewards earned by users are slightly higher on 'Sikka Pro' app as compared to 'Sikka' app, which could contribute to better user retention on 'Sikka Pro' app.

In terms of revenue, both apps have generated similar revenue, with 'Sikka Pro' app generating slightly higher revenue than 'Sikka' app.

Based on these insights, it can be concluded that 'Sikka Pro' app has better user engagement and experience, which could contribute to higher user retention in the long run. However, both apps have generated similar revenue, indicating that they are both profitable and successful.
